## Checking for minority groups
The following code will look at all of each society's data and find what constitutes a minority group for each principle value

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import random


# Path to the folder containing the CSV files
folder_path = '/home/ia23938/Documents/GitHub/ValueSystemsAggregation/data/society_data/norm_societies/'

# List to hold dataframes
dfs = []

# Load all CSV files into dataframes and append to the list
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        df = pd.read_csv(os.path.join(folder_path, filename))
        dfs.append(df)

# Concatenate all dataframes
all_data = pd.concat(dfs)

# Calculate the average for every column
average_data = all_data.mean()

average_data_df = average_data.to_frame().T
print(average_data_df)
```

```python
society_nums = [69, 39, 57, 42, 25, 85, 74, 84, 16, 21]

society_nums = list(range(1, 99))
# Path to the folder containing the CSV files
folder_path = '/home/ia23938/Documents/GitHub/ValueSystemsAggregation/data/society_data/norm_societies/'

dfs = []

for society_num in society_nums:
    df = pd.read_csv(os.path.join(folder_path, 'norm_society_'+str(society_num)+'.csv'))
    dfs.append(df)


# Concatenate all dataframes
all_data = pd.concat(dfs)

# Create a list of dataframes where each dataframe contains all rows for a unique agent_id
unique_agent_dfs = [all_data[all_data['agent_id'] == agent_id] for agent_id in all_data['agent_id'].unique()]

# Calculate the mean and standard deviation for all_data
mean_all_data = all_data.mean()
std_all_data = all_data.std()

unique_agent_dfs[0]
```

```python
for df in unique_agent_dfs:
    agent_data = df.set_index('agent_id')
    # Identify columns where agent_data is significantly different from all_data
    significant_columns = []
    for col in agent_data.columns:
        for row in agent_data[col]:
            if abs(row - mean_all_data[col]) > 2 * std_all_data[col]:
                print('row is: ', row)
                print('mean is: ', mean_all_data[col])
                print("found a significant row")
        if abs(agent_data[col].mean() - mean_all_data[col]) > 2 * std_all_data[col]:
            significant_columns.append(col)

    # Plot histograms for significant columns
    if significant_columns:
        agent_data[significant_columns].hist(figsize=(1.5, 1.5))
        plt.xlim(-1, 1)
        plt.show()
        print("FOUND! Agent number: ", df['agent_id'].iloc[0])
    #else:
```
