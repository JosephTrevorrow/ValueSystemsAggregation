```python
import numpy as np
import matplotlib.pyplot as plt
rng = np.random.default_rng()
agent_id = 1
rand_prefs = rng.normal(loc=0.8, scale=0.1, size=1000)
rand_prefs = np.clip(rand_prefs, 0, 1)
plt.hist(rand_prefs, bins=30, density=True)
plt.title('Normally Distributed Random Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.default_rng(seed=2024)
sample = rng.normal(loc=0.5, scale=0.1, size=2000)

plt.hist(sample, bins=30, density=True)
plt.title('Normally Distributed Random Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
